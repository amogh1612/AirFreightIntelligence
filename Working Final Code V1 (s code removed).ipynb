```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Bobby Co Worldwide Air Shipment Volume.csv")
df.head()
```

```python
raw_df = pd.read_csv("Bobby Co Worldwide Air Shipment Volume.csv")
df = raw_df.copy()
```

```python
## Finds largest lanes (top 10) 

lane_totals = df.groupby(['ORIGIN_BRANCH', 'DESTIN_BRANCH'])['WEIGHT_CHARGEABLE'].sum().reset_index()

largest_lanes = lane_totals.sort_values(by='WEIGHT_CHARGEABLE', ascending=False)

largest_df = largest_lanes.head(10)
largest_df
```

```python
# Create a list of (origin, destination) tuples

lane_list = list(largest_df[['ORIGIN_BRANCH', 'DESTIN_BRANCH']].itertuples(index=False, name=None))
lane_list
```

```python
## uses all lane combos in dictionary to extract rows from original df

for origin, destination in lane_list:
    lane_df = df[(df['ORIGIN_BRANCH'] == origin) & (df['DESTIN_BRANCH'] == destination)]    
    print(f"Lane: {origin} -> {destination}, Rows: {len(lane_df)}")

lane_dfs = {
    (origin, destination): df[(df['ORIGIN_BRANCH'] == origin) & (df['DESTIN_BRANCH'] == destination)].sort_values('LOCAL_EVENT_DATE')
    for origin, destination in lane_list
}

lane_dfs['PVG', 'AMS']
```

```python
## Filling empty date values for time series

lane_dfs_filled = {}

for origin, destination in lane_list:
    lane_df = df[(df['ORIGIN_BRANCH'] == origin) & (df['DESTIN_BRANCH'] == destination)].copy()
    
    # Make sure date is datetime
    lane_df['LOCAL_EVENT_DATE'] = pd.to_datetime(lane_df['LOCAL_EVENT_DATE'])
    
    # Create full date range for this lane
    full_range = pd.date_range(
        start=lane_df['LOCAL_EVENT_DATE'].min(),
        end=lane_df['LOCAL_EVENT_DATE'].max()
    )

    # Create full frame with all dates
    full_dates_df = pd.DataFrame({'LOCAL_EVENT_DATE': full_range})
    full_dates_df['ORIGIN_BRANCH'] = origin
    full_dates_df['DESTIN_BRANCH'] = destination

    # Merge to ensure every date is present
    filled_lane_df = full_dates_df.merge(
        lane_df,
        on=['ORIGIN_BRANCH', 'DESTIN_BRANCH', 'LOCAL_EVENT_DATE'],
        how='left'
    )

    # Fill missing weight with 1
    filled_lane_df['WEIGHT_CHARGEABLE'] = filled_lane_df['WEIGHT_CHARGEABLE'].fillna(1)

    # Save to dictionary for later use
    lane_dfs_filled[(origin, destination)] = filled_lane_df.sort_values('LOCAL_EVENT_DATE')
```

```python
lane_dfs_filled['PVG', 'AMS']
```

```python
from prophet import Prophet

forecasts = {}

for lane, lane_df in lane_dfs_filled.items():
    origin, destination = lane

    # Rename columns to Prophet format
    prophet_df = lane_df[['LOCAL_EVENT_DATE', 'WEIGHT_CHARGEABLE']].rename(
        columns={'LOCAL_EVENT_DATE': 'ds', 'WEIGHT_CHARGEABLE': 'y'}
    )

    # Initialize and fit the model
    model = Prophet(daily_seasonality=True)
    model.fit(prophet_df)

    # Create future dataframe
    future = model.make_future_dataframe(periods=30)
    forecast = model.predict(future)

    # Store results
    forecasts[lane] = {
        'model': model,
        'forecast': forecast
    }

    print(f" Complete: {origin} -> {destination}")
```

```python
lane = ('PVG', 'ATL')
model = forecasts[lane]['model']
forecast = forecasts[lane]['forecast']

fig = model.plot(forecast)
plt.title(f"Forecast for lane {lane[0]} → {lane[1]}")
plt.show()
```
