```python
import pandas as pd

df = pd.read_csv("Capacity Prediction at co Level.csv")

# adjust column names if needed
y_true = df["Observed Chargeable Weight"]
y_pred = df["Predicted Chargeable Weight considering Holidays"]
y_pred1 = df["Predicted Chargeable Weight"]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

def smape(y_true, y_pred):
    return np.mean(
        2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred))
    ) * 100

def evaluate_model(name, y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    smape_val = smape(y_true, y_pred)
    
    print(f"\n{name}")
    print("MAE:", mae)
    print("RMSE:", rmse)
    print("sMAPE:", smape_val)

# Evaluate both
evaluate_model("With Holidays", y_true, y_pred)
evaluate_model("Without Holidays", y_true, y_pred1)
```

```python
import numpy as np

tolerance = 0.35  # 15%

# Avoid division by zero by adding a small epsilon
epsilon = 1e-8

percent_error = np.abs(y_pred - y_true) / (np.abs(y_true) + epsilon)

# Optional: remove rows where actual is extremely small (recommended)
mask = np.abs(y_true) > 1000   # adjust threshold if needed
filtered_error = percent_error[mask]

accuracy = (filtered_error <= tolerance).mean()

print(f"Accuracy within {tolerance*100:.0f}%:", accuracy)
print(f"Rows used: {mask.sum()} / {len(df)}")
```
