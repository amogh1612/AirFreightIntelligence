```python
## Filling empty date values for time series

lane_dfs_filled = {}

for origin, destination in lane_list:
    lane_df = df[(df['ORIGIN_BRANCH'] == origin) & (df['DESTIN_BRANCH'] == destination)].copy()
    
    # Make sure date is datetime
    lane_df['LOCAL_EVENT_DATE'] = pd.to_datetime(lane_df['LOCAL_EVENT_DATE'])
    
    # Create full date range for this lane
    full_range = pd.date_range(
        start=lane_df['LOCAL_EVENT_DATE'].min(),
        end=lane_df['LOCAL_EVENT_DATE'].max()
    )

    # Create full frame with all dates
    full_dates_df = pd.DataFrame({'LOCAL_EVENT_DATE': full_range})
    full_dates_df['ORIGIN_BRANCH'] = origin
    full_dates_df['DESTIN_BRANCH'] = destination

    # Merge to ensure every date is present
    filled_lane_df = full_dates_df.merge(
        lane_df,
        on=['ORIGIN_BRANCH', 'DESTIN_BRANCH', 'LOCAL_EVENT_DATE'],
        how='left'
    )

    # Fill missing weight with 1
    filled_lane_df['WEIGHT_CHARGEABLE'] = filled_lane_df['WEIGHT_CHARGEABLE'].fillna(1)

    # Save to dictionary for later use
    lane_dfs_filled[(origin, destination)] = filled_lane_df.sort_values('LOCAL_EVENT_DATE')
```


# Validation + Preprocessing + Model Comparison (Top 5 Lanes)

This section strengthens the forecasting workflow for a machine learning class setting by adding:

- **Data cleansing / preprocessing**
- **Top 5 lane selection**
- **Distribution checks and `log1p` transformation**
- **Time-aware train/test split**
- **Rolling-window validation**
- **Comparison of two common time series models: Prophet and SARIMAX**
- **Inverse transformation with `expm1` so predictions are reported in original weight units**

**Important notes**
- Because this is time series data, we do **not** shuffle the rows.
- We now keep missing dates as **0 volume**, then apply `log1p`. This is more principled than replacing missing days with 1 before modeling.
- The code below is written so it can run off the same shipment dataset used earlier in the notebook.


```python

# ---------------------------------------------------------
# Section Setup
# ---------------------------------------------------------
import warnings
warnings.filterwarnings("ignore")

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import TimeSeriesSplit
from prophet import Prophet
from statsmodels.tsa.statespace.sarimax import SARIMAX

pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 140)

REQUIRED_COLUMNS = ["ORIGIN_BRANCH", "DESTIN_BRANCH", "LOCAL_EVENT_DATE", "WEIGHT_CHARGEABLE"]

def load_shipment_data():
    '''
    Reuse raw_df if it already exists from earlier cells.
    Otherwise, attempt to read the original CSV file.
    '''
    if "raw_df" in globals():
        data = raw_df.copy()
    elif os.path.exists("Bobby Co Worldwide Air Shipment Volume.csv"):
        data = pd.read_csv("Bobby Co Worldwide Air Shipment Volume.csv")
    else:
        raise FileNotFoundError(
            "Could not find the shipment CSV. Make sure "
            "'Bobby Co Worldwide Air Shipment Volume.csv' is available."
        )

    missing = [c for c in REQUIRED_COLUMNS if c not in data.columns]
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    return data

df_ts = load_shipment_data()
df_ts.head()
```

```python

# ---------------------------------------------------------
# Data Cleansing / Preprocessing
# ---------------------------------------------------------
df_ts = df_ts.copy()
df_ts = df_ts[REQUIRED_COLUMNS].copy()

df_ts["LOCAL_EVENT_DATE"] = pd.to_datetime(df_ts["LOCAL_EVENT_DATE"], errors="coerce")

df_ts["WEIGHT_CHARGEABLE"] = (
    df_ts["WEIGHT_CHARGEABLE"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace("$", "", regex=False)
    .str.strip()
)
df_ts["WEIGHT_CHARGEABLE"] = pd.to_numeric(df_ts["WEIGHT_CHARGEABLE"], errors="coerce")

df_ts = df_ts.dropna(subset=["LOCAL_EVENT_DATE", "WEIGHT_CHARGEABLE", "ORIGIN_BRANCH", "DESTIN_BRANCH"]).copy()
df_ts = df_ts[df_ts["WEIGHT_CHARGEABLE"] >= 0].copy()

df_ts["ORIGIN_BRANCH"] = df_ts["ORIGIN_BRANCH"].astype(str).str.strip().str.upper()
df_ts["DESTIN_BRANCH"] = df_ts["DESTIN_BRANCH"].astype(str).str.strip().str.upper()

df_ts["LANE"] = df_ts["ORIGIN_BRANCH"] + " -> " + df_ts["DESTIN_BRANCH"]

print("Rows after cleansing:", len(df_ts))
print("Date range:", df_ts["LOCAL_EVENT_DATE"].min(), "to", df_ts["LOCAL_EVENT_DATE"].max())
df_ts.head()
```

```python

# ---------------------------------------------------------
# Focus on Top 5 Lanes by Total Volume
# ---------------------------------------------------------
top5_lanes = (
    df_ts.groupby("LANE", as_index=False)["WEIGHT_CHARGEABLE"]
    .sum()
    .sort_values("WEIGHT_CHARGEABLE", ascending=False)
    .head(5)
)

top5_lanes
```

```python

# ---------------------------------------------------------
# Build Daily Time Series Per Lane
# ---------------------------------------------------------
def prepare_lane_daily_series(data, lane):
    '''
    Aggregate a lane to daily level and fill missing dates with 0.
    This is important before using TimeSeriesSplit or seasonality models.
    '''
    lane_df = data[data["LANE"] == lane].copy()
    lane_daily = (
        lane_df.groupby("LOCAL_EVENT_DATE", as_index=False)["WEIGHT_CHARGEABLE"]
        .sum()
        .sort_values("LOCAL_EVENT_DATE")
        .rename(columns={"LOCAL_EVENT_DATE": "ds", "WEIGHT_CHARGEABLE": "y"})
    )

    full_dates = pd.date_range(lane_daily["ds"].min(), lane_daily["ds"].max(), freq="D")
    lane_daily = (
        lane_daily.set_index("ds")
        .reindex(full_dates)
        .rename_axis("ds")
        .reset_index()
    )
    lane_daily["y"] = lane_daily["y"].fillna(0.0)
    lane_daily["is_sunday"] = (lane_daily["ds"].dt.weekday == 6).astype(int)
    lane_daily["y_log1p"] = np.log1p(lane_daily["y"])

    return lane_daily

lane_series = {
    lane: prepare_lane_daily_series(df_ts, lane)
    for lane in top5_lanes["LANE"]
}

example_lane = top5_lanes["LANE"].iloc[0]
lane_series[example_lane].head()
```

```python

# ---------------------------------------------------------
# Check Distribution: Raw vs log1p
# ---------------------------------------------------------
lane_to_plot = top5_lanes["LANE"].iloc[0]
plot_df = lane_series[lane_to_plot]

fig, axes = plt.subplots(1, 2, figsize=(14, 4))
axes[0].hist(plot_df["y"], bins=30)
axes[0].set_title(f"Raw Distribution\n{lane_to_plot}")
axes[0].set_xlabel("Daily Weight")
axes[0].set_ylabel("Count")

axes[1].hist(plot_df["y_log1p"], bins=30)
axes[1].set_title(f"log1p Distribution\n{lane_to_plot}")
axes[1].set_xlabel("log1p(Daily Weight)")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()

print("Skewness (raw):", round(plot_df["y"].skew(), 3))
print("Skewness (log1p):", round(plot_df["y_log1p"].skew(), 3))
```

```python

# ---------------------------------------------------------
# Metrics
# ---------------------------------------------------------
def rmse(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def mae(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    return np.mean(np.abs(y_true - y_pred))

def mape(y_true, y_pred, eps=1e-8):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = y_true > eps
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

def smape(y_true, y_pred, eps=1e-8):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    denom = np.abs(y_true) + np.abs(y_pred)
    mask = denom > eps
    if mask.sum() == 0:
        return np.nan
    return np.mean(2.0 * np.abs(y_true[mask] - y_pred[mask]) / denom[mask]) * 100

def evaluate_predictions(y_true, y_pred):
    return {
        "RMSE": rmse(y_true, y_pred),
        "MAE": mae(y_true, y_pred),
        "MAPE_pct": mape(y_true, y_pred),
        "SMAPE_pct": smape(y_true, y_pred),
    }
```

```python

# ---------------------------------------------------------
# Model Helpers
# ---------------------------------------------------------
def fit_predict_prophet(train_df, test_df):
    '''
    Train Prophet on log1p-transformed target and predict on the test dates.
    We inverse-transform with expm1 at the end.
    '''
    prophet_train = train_df[["ds", "y_log1p", "is_sunday"]].rename(columns={"y_log1p": "y"}).copy()

    model = Prophet(
        daily_seasonality=False,
        weekly_seasonality=True,
        yearly_seasonality=True
    )
    model.add_regressor("is_sunday")
    model.fit(prophet_train)

    future = test_df[["ds", "is_sunday"]].copy()
    forecast = model.predict(future)

    pred_log = forecast["yhat"].values
    pred = np.expm1(pred_log)
    pred = np.maximum(pred, 0)

    return pred, model, forecast

def fit_predict_sarimax(train_df, test_df):
    '''
    Simple SARIMAX baseline on log1p-transformed target.
    Seasonal period 7 captures weekly seasonality on daily data.
    '''
    model = SARIMAX(
        train_df["y_log1p"],
        exog=train_df[["is_sunday"]],
        order=(1, 1, 1),
        seasonal_order=(1, 1, 1, 7),
        enforce_stationarity=False,
        enforce_invertibility=False
    )

    results = model.fit(disp=False)

    forecast_log = results.forecast(
        steps=len(test_df),
        exog=test_df[["is_sunday"]]
    )

    pred = np.expm1(np.asarray(forecast_log))
    pred = np.maximum(pred, 0)

    return pred, results
```

```python

# ---------------------------------------------------------
# Time-Aware Holdout Split + Rolling Window Validation
# ---------------------------------------------------------
def get_time_holdout_split(lane_df, test_days=30):
    '''
    Keep the last test_days for final evaluation.
    If the series is short, use about 20% as holdout with a sensible minimum.
    '''
    n = len(lane_df)
    dynamic_test = min(test_days, max(14, int(round(n * 0.2))))
    split_idx = n - dynamic_test
    train_df = lane_df.iloc[:split_idx].copy()
    test_df = lane_df.iloc[split_idx:].copy()
    return train_df, test_df

def rolling_cv_compare_models(train_df, n_splits=3, test_size=21):
    '''
    Rolling validation inside the training portion only.
    '''
    results = []

    effective_test_size = min(test_size, max(7, len(train_df) // (n_splits + 2)))
    tscv = TimeSeriesSplit(n_splits=n_splits, test_size=effective_test_size)

    for fold, (tr_idx, val_idx) in enumerate(tscv.split(train_df), start=1):
        tr = train_df.iloc[tr_idx].copy()
        val = train_df.iloc[val_idx].copy()

        prophet_pred, _, _ = fit_predict_prophet(tr, val)
        prophet_metrics = evaluate_predictions(val["y"].values, prophet_pred)
        prophet_metrics.update({"Model": "Prophet", "Fold": fold})
        results.append(prophet_metrics)

        sarimax_pred, _ = fit_predict_sarimax(tr, val)
        sarimax_metrics = evaluate_predictions(val["y"].values, sarimax_pred)
        sarimax_metrics.update({"Model": "SARIMAX", "Fold": fold})
        results.append(sarimax_metrics)

    return pd.DataFrame(results)
```

```python

# ---------------------------------------------------------
# Run Validation + Final Holdout Evaluation for Top 5 Lanes
# ---------------------------------------------------------
all_cv_results = []
all_holdout_results = []
forecast_outputs = {}

for lane, lane_df in lane_series.items():
    if len(lane_df) < 60:
        print(f"Skipping {lane}: not enough daily history ({len(lane_df)} rows)")
        continue

    train_df, test_df = get_time_holdout_split(lane_df, test_days=30)

    cv_results = rolling_cv_compare_models(train_df, n_splits=3, test_size=21)
    cv_results["Lane"] = lane
    all_cv_results.append(cv_results)

    cv_summary = (
        cv_results.groupby("Model", as_index=False)[["RMSE", "MAE", "MAPE_pct", "SMAPE_pct"]]
        .mean()
        .sort_values("SMAPE_pct")
    )

    best_model = cv_summary.iloc[0]["Model"]

    if best_model == "Prophet":
        holdout_pred, fitted_model, raw_forecast = fit_predict_prophet(train_df, test_df)
    else:
        holdout_pred, fitted_model = fit_predict_sarimax(train_df, test_df)
        raw_forecast = None

    holdout_metrics = evaluate_predictions(test_df["y"].values, holdout_pred)
    holdout_metrics.update({
        "Lane": lane,
        "Best_Model_From_CV": best_model,
        "Train_Rows": len(train_df),
        "Test_Rows": len(test_df),
    })
    all_holdout_results.append(holdout_metrics)

    forecast_outputs[lane] = {
        "train_df": train_df,
        "test_df": test_df,
        "pred": holdout_pred,
        "best_model": best_model,
        "fitted_model": fitted_model,
        "raw_forecast": raw_forecast,
    }

cv_results_df = pd.concat(all_cv_results, ignore_index=True) if all_cv_results else pd.DataFrame()
holdout_results_df = pd.DataFrame(all_holdout_results)

print("Cross-validation results:")
display(cv_results_df.head())

print("\nFinal holdout results:")
display(holdout_results_df.sort_values("SMAPE_pct"))
```

```python

# ---------------------------------------------------------
# Aggregate CV Summary by Lane and Model
# ---------------------------------------------------------
if not cv_results_df.empty:
    cv_summary_df = (
        cv_results_df
        .groupby(["Lane", "Model"], as_index=False)[["RMSE", "MAE", "MAPE_pct", "SMAPE_pct"]]
        .mean()
        .sort_values(["Lane", "SMAPE_pct"])
    )
    display(cv_summary_df)
else:
    print("No CV results available.")
```

```python

# ---------------------------------------------------------
# Plot Final Holdout Forecast for One Lane
# ---------------------------------------------------------
lane_to_view = holdout_results_df.sort_values("SMAPE_pct").iloc[0]["Lane"]
out = forecast_outputs[lane_to_view]

plt.figure(figsize=(14, 5))
plt.plot(out["train_df"]["ds"], out["train_df"]["y"], label="Train Actual")
plt.plot(out["test_df"]["ds"], out["test_df"]["y"], label="Test Actual")
plt.plot(out["test_df"]["ds"], out["pred"], label=f"Predicted ({out['best_model']})")
plt.title(f"Final Holdout Forecast: {lane_to_view}")
plt.xlabel("Date")
plt.ylabel("Chargeable Weight")
plt.legend()
plt.show()

print("Best model:", out["best_model"])
display(holdout_results_df[holdout_results_df["Lane"] == lane_to_view])
```


## Suggested write-up points for the report / proposal

You can use the results from the section above to discuss:

1. **Why preprocessing matters**  
   The raw shipment data should be cleaned carefully: dates must be parsed correctly, the target must be numeric, and missing daily dates should be filled before modeling.

2. **Why `log1p` is useful here**  
   Daily freight volume is usually right-skewed. A `log1p` transform often makes the distribution less skewed and stabilizes variance. Since `log1p(0)` is defined, it is a natural choice when some days have zero volume.

3. **Why normal train/test splitting is not enough**  
   Since this is time series data, the model must be evaluated using time-aware splits. Random shuffling would leak future information into the training process.

4. **Why rolling validation is better for this problem**  
   Rolling-window validation better reflects how the model would behave in practice: train on earlier history, then forecast the next unseen block of time.

5. **Why compare Prophet and SARIMAX**  
   - **Prophet** is convenient for trend + seasonality and works well when you expect interpretable weekly/yearly patterns.
   - **SARIMAX** is a strong statistical benchmark for autoregressive seasonal structure.
   Comparing both gives a more defensible modeling story for a class project.

6. **Why inverse transformation is required**  
   If the model is trained on the transformed target, the predictions must be converted back with `expm1()` before reporting forecasted weights. Otherwise, the reported values are not in the original business units.

7. **How to connect this to holidays / seasonality**  
   Seasonal structure may already absorb patterns tied to recurring events like Chinese New Year. A next extension would be adding explicit holiday calendars for lanes that are strongly affected by specific regional holidays.
